```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x,y,'red')
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"red")

plt.subplot(1,2,2)
plt.plot(y,x,'blue') 
```

```python
# OBJECT ORIENTED
fig = plt.figure()

# array is the location of canvas
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,0.8,0.8])
axes2 = fig.add_axes([0.1,0.4,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Outer Plot")

axes2.plot(y,x,'r')
axes2.set_title("Inner Plot")
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x,'r')

#to care overlapping
plt.tight_layout()
```

## Figure size aspect ratio and DPI

```python
fig = plt.figure(figsize=(5,3))

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(20,5))

axes[0].plot(x,y)
axes[1].plot(y,x,'r')
```

```python
fig.savefig("My_fig.png",dpi=200)
```

```python
fig = plt.figure(figsize=(5,3))

axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2, label="X squared" , linewidth=4, alpha=0.5)
axes.plot(x,x**3, label="X cubed", linestyle="--")
axes.legend(loc=0)
```

```python
fig = plt.figure(figsize=(10,5))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color="purple" ,lw=2, marker="o",markersize=10)
```

```python
fig = plt.figure(figsize=(10,5))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color="purple" ,lw=2)
axes.set_xlim([0,1])
axes.set_ylim([0,1])
```
